Creating a data frame from MarketWatch data

[Website](https://www.marketwatch.com/investing/cryptocurrency)

```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

Import and clean the BTC data set

```python
# import and clean the BTC data frame

df_btc = pd.read_csv('/content/BTCUSD.csv') # read in BTC data

df_btc['Name'] = 'BTC' # set the name
df_btc['Close'] = df_btc['Close'].astype(str).str.replace(',', '').astype(float) # convert to float
df_btc['Close'] = df_btc['Close'].astype(int) # convert to int
df_btc['Change'] = df_btc['Close'].diff() # find the difference
df_btc[['Open', 'High', 'Low', 'Close']] = df_btc[['Open', 'High', 'Low', 'Close']].astype(object) #change to objects\
df_btc['%'] = (df_btc['Change'] / df_btc['Close']) * 100 # calculate the % change
df_btc['%'] = pd.to_numeric(df_btc['%'], errors='coerce')  # convert to numeric, replace non-numeric with NaN
df_btc['%'] = df_btc['%'].round(2) # round to 2 decimals
df_btc['Change'] = df_btc['Change'].astype(object) # convert to object
df_btc['%'] = df_btc['%'].astype(object) # convert to object
print(df_btc.dtypes) # show data types
df_btc.head()
```

Import and clean the ETH data set

```python
# import and clean the ETH data frame

df_eth = pd.read_csv('/content/ETHUSD.csv') # read in ETH data

df_eth['Name'] = 'ETH' # set the name
df_eth['Close'] = df_eth['Close'].astype(str).str.replace(',', '').astype(float) # convert to float
df_eth['Close'] = df_eth['Close'].astype(int) # convert to int
df_eth['Change'] = df_eth['Close'].diff() # find the difference
df_eth[['Open', 'High', 'Low', 'Close']] = df_eth[['Open', 'High', 'Low', 'Close']].astype(object) #change to objects\
df_eth['%'] = (df_eth['Change'] / df_eth['Close']) * 100 # calculate the % change
df_eth['%'] = pd.to_numeric(df_eth['%'], errors='coerce')  # convert to numeric, replace non-numeric with NaN
df_eth['%'] = df_eth['%'].round(2) # round to 2 decimals
df_eth['Change'] = df_eth['Change'].astype(object) # convert to object
df_eth['%'] = df_eth['%'].astype(object) # convert to object
print(df_eth.dtypes) # show data types
df_eth.head()
```

Import and clean the SOL data set

```python
# import and clean the SOL data frame

df_sol = pd.read_csv('/content/SOLUSD.csv') # read in SOL data

df_sol['Name'] = 'SOL' # set the name
df_sol['Close'] = df_sol['Close'].astype(str).str.replace(',', '').astype(float) # convert to float
df_sol['Close'] = df_sol['Close'].astype(int) # convert to int
df_sol['Change'] = df_sol['Close'].diff() # find the difference
df_sol[['Open', 'High', 'Low', 'Close']] = df_sol[['Open', 'High', 'Low', 'Close']].astype(object) #change to objects\
df_sol['%'] = (df_sol['Change'] / df_sol['Close']) * 100 # calculate the % change
df_sol['%'] = pd.to_numeric(df_sol['%'], errors='coerce')  # convert to numeric, replace non-numeric with NaN
df_sol['%'] = df_sol['%'].round(2) # round to 2 decimals
df_sol['Change'] = df_sol['Change'].astype(object) # convert to object
df_sol['%'] = df_sol['%'].astype(object) # convert to object
print(df_sol.dtypes) # show data types
df_sol.head()
```

Import and clean the ADA data set

```python
# import and clean the ADA data frame

df_ada = pd.read_csv('/content/ADAUSD.csv') # read in ADA data

df_ada['Name'] = 'ADA' # add the name
df_ada['Close'] = df_ada['Close'].round(4)  # round to 2 decimal places
df_ada['Change'] = df_ada['Close'].diff()  # get the difference
df_ada = df_ada.astype(object) # convert all to objects
print(df_ada.dtypes)
df_ada.head()
```

Import and clean the XRP data set

```python
# import and clean the XRP data frame

df_xrp = pd.read_csv('/content/XRPUSD.csv') # read in XRP data

df_xrp['Name'] = 'ADA' # add the name
df_xrp['Close'] = df_xrp['Close'].round(4)  # round to 2 decimal places
df_xrp['Change'] = df_xrp['Close'].diff()  # get the difference
df_xrp = df_xrp.astype(object) # convert all to objects
print(df_xrp.dtypes)
df_xrp.head()
```

Combine to create the new data set

```python
# combine all the data frames to 1 data frame

df_all = pd.concat([df_btc, df_eth, df_sol, df_ada, df_xrp])
df_all = df_all[['Date', 'Name', 'Open', 'High', 'Low', 'Close', 'Change', '%']] # change the column order
print(f'Tokens =  {df_all["Name"].unique()}') # show the tokens used
print(f"{df_all['Date'].min()} - {df_all['Date'].max()}") # print the date range in the data set
print(f'{df_all.shape}') # show the data frame shape
df_all.head()
```

```python
# download the new data frame

df_all.to_csv('ALLDATA.csv', index=False)
files.download('ALLDATA.csv')
```

```python
# get the ATH and ATL for a chosen token

name = 'BTC' # choose the token

max_high_row = df_all[(df_all['Name'] == name) & (df_all['High'] == df_all[df_all['Name'] == name]['High'].max())] # find the max value in High
max_high_date = max_high_row['Date'].values[0] # get the date for the High
max_high_value = max_high_row['High'].values[0] # get the value for the High
min_low_row = df_all[(df_all['Name'] == name) & (df_all['Low'] == df_all[df_all['Name'] == name]['Low'].min())] # find the max value in Low
min_low_date = min_low_row['Date'].values[0] # get the date for the low
min_low_value = min_low_row['Low'].values[0] # get the value for the low

print(f"Date of max high for BTC: {max_high_date}")
print(f"Max high value for BTC: {max_high_value}")
print(f'-')
print(f"Date of min low for BTC: {min_low_date}")
print(f"Min low value for BTC: {min_low_value}")
```

```python
# get the largest and lowest daily change for a chosen token

name = 'BTC' # choose the token

largest_change_row = df_all[(df_all['Name'] == name) & (df_all['Change'] == df_all[df_all['Name'] == name]['Change'].max())] # find the max value in High
largest_change_date = largest_change_row['Date'].values[0] # get the date for the High
largest_change_value = largest_change_row['Change'].values[0] # get the value for the High
lowest_change_row = df_all[(df_all['Name'] == name) & (df_all['Change'] == df_all[df_all['Name'] == name]['Change'].min())] # find the max value in Low
lowest_change_date = lowest_change_row['Date'].values[0] # get the date for the low
lowest_change_value = lowest_change_row['Change'].values[0] # get the value for the low

print(f"Date of largest change for BTC: {largest_change_date}")
print(f"Largest change value for BTC: {largest_change_value}")
print(f'-')
print(f"Date of lowest change for BTC: {lowest_change_date}")
print(f"Lowest change value for BTC: {lowest_change_value}")
```

```python
# display BTC in a line chart, x axis is the date, y axis is the close

df_btc_only = df_all[df_all['Name'] == 'BTC'] # filter for the token
df_btc_only['Date'] = pd.to_datetime(df_btc_only['Date']) # convert the date column to datetime objects

# create the line chart
plt.figure(figsize=(10, 6))
plt.plot(df_btc_only['Date'], df_btc_only['Close'])
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('BTC Closing Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
