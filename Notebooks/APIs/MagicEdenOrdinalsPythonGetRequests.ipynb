```python
import requests
```

```python
### Using Python and Get Requests to retrieve data from the MagicEden Ordinals API. ###
```

```python
# get a collections info
user_input = input("Enter a collection name: ")

def get_collection_info(collection_symbol):
  url = f"https://api-mainnet.magiceden.dev/v2/ord/btc/collections/{collection_symbol}"
  headers = {
    "User-Agent": "Your_User_Agent"  # Replace with your user agent
  }

  try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    data = response.json()
    return data
  except requests.exceptions.RequestException as e:
    print(f"Error fetching collection info: {e}")
    return None

if __name__ == "__main__":
  collection_symbol = user_input # Replace with the desired symbol
  collection_info = get_collection_info(collection_symbol)
  if collection_info:
    print(collection_info)
  else:
    print("Error: Could not retrieve collection information.")
```

```python
# get a collections stats
user_input = input("Enter a collection name: ")

def get_collection_stats(collection_symbol):
  url = f"https://api-mainnet.magiceden.dev/v2/collections/{collection_symbol}/stats"
  headers = {
    "User-Agent": "Your_User_Agent"  # Replace with your user agent
  }

  try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    data = response.json()
    return data
  except requests.exceptions.RequestException as e:
    print(f"Error fetching collection stats: {e}")
    return None

if __name__ == "__main__":
  collection_symbol = user_input  # Replace with the desired symbol
  stats = get_collection_stats(collection_symbol)
  if stats:
    print(stats)
  else:
    print("Error: Could not retrieve collection stats.")
```
