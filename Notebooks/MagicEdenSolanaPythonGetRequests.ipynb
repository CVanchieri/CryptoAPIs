```python
# get the collections stats
user_input = input("Enter a SOL collection name: ")

def get_collection_stats(collection_symbol):
  url = f"https://api-mainnet.magiceden.dev/v2/collections/{collection_symbol}/stats"
  headers = {
    "User-Agent": "Your_User_Agent"  # Replace with your user agent
  }

  try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    data = response.json()
    return data
  except requests.exceptions.RequestException as e:
    print(f"Error fetching collection stats: {e}")
    return None

if __name__ == "__main__":
  collection_symbol = user_input # Replace with the desired collection symbol
  stats = get_collection_stats(collection_symbol)
  if stats:
    print(stats)
```

```python
# get tokens owned in wallet
user_input = input("Enter a SOL wallet address: ")

def get_wallet_tokens(wallet_address):
  url = f"https://api-mainnet.magiceden.dev/v2/wallets/{wallet_address}/tokens"
  headers = {
    "User-Agent": "Your_User_Agent"  # Replace with your user agent
  }

  try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    data = response.json()
    return data
  except requests.exceptions.RequestException as e:
    print(f"Error fetching wallet tokens: {e}")
    return None

if __name__ == "__main__":
  wallet_address = user_input  # Replace with the desired wallet address
  owned_tokens = get_wallet_tokens(wallet_address)
  if owned_tokens:
    print(owned_tokens)
  else:
    print("Error: Could not retrieve wallet tokens.")
```
